# Reading the Json Files in pandas df

```python
#file to  json  with different formats with index
df.to_json(orient='index')
```

```python
#file to save as json file from json data
df.to_json('index of wine.json')
```

```python
#file to  json  with different formats with columns 
df.to_json(orient='columns')
```

```python
#file to  json  with different formats with records
df.to_json(orient='records')
```

```python
#file to  json  with different formats with split
df.to_json(orient='split')
```

```python
#file to  json  with different formats with values
df.to_json(orient='values')
```

# Reading HTML contents in pandas 

```python
url_mcc = 'https://en.wikipedia.org/wiki/Mobile_country_code'
df_html = pd.read_html(url_mcc,match='Country',header=0)
```

```python
df_html[0]
```

```python
#saving the file to html in pandas 
df_html[0].to_html('mcc.html')
```
